```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"
sd_pipeline = StableDiffusionPipeline.from_pretrained(model_id,torch_dtype=torch.float16)

sd_pipeline = sd_pipeline.to("cuda")
```

```python
def get_completion_sd(prompt):
  negative_prompt="""
  simple background,duplicate,low quality,lowest quality,bad anatomy,bad proportion,extra digits,
  lowres,username,artist name,error,duplicate,watermark,signature,text,extra digits,worst quality,
  jpeg artifacts,blurry
  """
  return sd_pipeline(prompt,negative_prompt=negative_prompt).images[0]
```

```python
#lets prompt

prompt = "astronaut,riding a horse,on mars,human colony"
#promt = "children,playing in disneyland,view from a distance"

# prompt = "baby llama,wearing red muffler,grazing,open field,sunset"



print(prompt)
sd_image = get_completion_sd(prompt)
# sd_image.save("./llama.jpg")
sd_image.save("./astronaut.jpg")
```

```python
import gradio as gr
```

```python
def get_completion(prompt):
  negative_prompt="""
  simple background,duplicate,high quality,highest quality,good anatomy,good proportion,extra digits,
  lowres,username,artist name,error,duplicate,watermark,signature,text,extra digits,worst quality,
  jpeg artifacts
  """
  return sd_pipeline(prompt,negative_prompt=negative_prompt,).images[0]

# def generate(prompt):
#   output = get_completion_sd(prompt)
#   return output

genai_app = gr.Interface(fn=get_completion,
                         inputs = [gr.Textbox(label="Your Prompt")],
                         outputs = [gr.Image(label="Result")],
                         title = "Generate Image as you want!",
                         description = "Generate any image with Stable Diffusion",
                         allow_flagging = "never",
                         examples = ["astronaut,riding a horse,on mars",
                                     "cargo ship,flying,in space"]
                         )
genai_app.launch()
```
